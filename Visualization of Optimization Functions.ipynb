```python
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.colors import LogNorm
from mpl_toolkits.mplot3d import Axes3D
```

```python
mlt.rcParams['axes.titlesize'] = 25
mlt.rcParams['axes.spines.top'] = False
mlt.rcParams['axes.spines.right'] = False
mlt.rcParams['figure.figsize'] = (12, 12)
```

```python
def xy(xmin, xmax, ymin, ymax, xstep, ystep):
    x = np.arange(xmin, xmax + xstep, xstep)
    y = np.arange(ymin, ymax + ystep, ystep)
    x, y = np.meshgrid(x, y)
    return x, y
```

```python
# Beale function 
def baeleFunction(x, y):
    f_1 = (1.5 - x + x * y) ** 2 + (2.25 - x + x * y ** 2) ** 2 + (2.625 - x + x * y ** 3) ** 2  
    return f_1

# Goldstein-price function
def goldsteinPriceFunction(x, y):
    f_2 = (1 + (x + y + 1) ** 2 * (19 - 14 * x + 3 * x ** 2 - 14 * y + 6 * x * y + 3 * y ** 2)) \
            * (30 + (2 * x - 3 * y) ** 2 * (18 - 32 * x + 12 * x ** 2 + 48 * y - 36 * x * y + 27 * y ** 2))
    return f_2
    
# Booth function
def boothFunction(x, y):
    f_3 = (x + 2 * y - 7) ** 2 + (2 * x + y - 5) ** 2
    return f_3

# Matyas function
def matyasFunction(x, y):
    f_4 = 0.26 * (x ** 2 + y ** 2) - 0.48 * x * y
    return f_4

# Levi function
def leviFunction(x, y):
    f_5 = np.sin(3 * np.pi * x) + (x - 1) ** 2 * (1 + np.sin(3 * np.pi * y) ** 2) \
            + (y - 1) ** 2 * (1 + np.sin(2 * np.pi * y) ** 2)
    return f_5

# Schaffer N2 Function
def schafferFunction(x, y):
    f_6 = 0.5 + (np.sin(x ** 2 - y ** 2)) ** 2 / (1 + 0.001 * (x ** 2 + y ** 2)) ** 2 
    return f_6

# Bukin N6 Function
def bukinFunction(x, y):
    f_7 = 100 * (np.sqrt(np.abs(y - 0.01 * x ** 2))) + 0.01 * np.abs(x + 10)
    return f_7
```

```python
def findGlobalMinimum(x, y, func):
    indx = np.unravel_index(func.argmin(), func.shape)
    x = x[indx[0]][indx[1]]
    y = y[indx[0]][indx[1]]
    point = func[indx[0]][indx[1]]
    coords = (x, y)
    return coords, point
```

```python
def plotFunction(x, y, func, constraints, coords, g_min, title=''):
    fig = plt.figure()
    ax = plt.axes(projection='3d', elev=40, azim=-50)
    ax.plot_surface(x, y, func, norm=LogNorm(), rstride=1, cstride=1, cmap=cm.jet, edgecolor='none', alpha=0.7)
    ax.plot(coords[0], coords[1], 'r*', markersize=10)
    ax.set_title(title)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.set_xlim((constraints['xmin'], constraints['xmax']))
    ax.set_ylim((constraints['ymin'], constraints['ymax']))
    plt.show()
    print(f'Global minimum: {g_min}')
```

```python
# Beale function
xmin, xmax, xstep = -4.5, 4.5, 0.1
ymin, ymax, ystep = -4.5, 4.5, 0.1
x, y = xy(xmin, xmax, ymin, ymax, xstep, ystep)
f_1 = baeleFunction(x, y)
coords, g_min = findGlobalMinimum(x, y, f_1)


constraints = {
    'xmin': xmin,
    'xmax': xmax,
    'ymin': ymin,
    'ymax': ymax,
}

plotFunction(x, y, f_1, constraints, coords, g_min, title='Beale Optimization Function')
```

```python
# Goldstein-price function
xmin, xmax, xstep = -5, 5, 0.1
ymin, ymax, ystep = -5, 5, 0.1
x, y = xy(xmin, xmax, ymin, ymax, xstep, ystep)
f_2 = goldsteinPriceFunction(x, y)
coords, g_min = findGlobalMinimum(x, y, f_2)

constraints = {
    'xmin': xmin,
    'xmax': xmax,
    'ymin': ymin,
    'ymax': ymax,
}

plotFunction(x, y, f_2, constraints, coords, g_min, title='Goldstein-Price Optimization Function')
```
