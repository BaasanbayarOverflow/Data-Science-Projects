```python
library(ggplot2)
library(dplyr)
library(readr)
```

```python
spotify_data_by_artists = read_csv("data/data_by_artist_o.csv")
sorted_data_by_popularity_artists <- spotify_data_by_artists %>%
                                    arrange(desc(popularity)) 
head(sorted_data_by_popularity_artists, 5)
tail(sorted_data_by_popularity_artists, 5)
```

```python
spotify_data_by_genres = read_csv("data/data_by_genres_o.csv")
sorted_data_by_popularity_genres <- spotify_data_by_genres %>%
                                        arrange(desc(popularity))
head(sorted_data_by_popularity_genres, 5)
tail(sorted_data_by_popularity_genres, 5)
```

```python
spotify_data_by_year = read.csv("data/data_by_year_o.csv")
sorted_data_by_popularity_year <- spotify_data_by_year %>%
                                        arrange(desc(popularity))
head(sorted_data_by_popularity_year, 5)
tail(sorted_data_by_popularity_year, 5)
```

```python
spotify_artists = read_csv("data/artists.csv")
sorted_data_artists_by_popularity <- spotify_artists %>% arrange(desc(popularity))
sorted_data_artists_by_followers <- spotify_artists %>% arrange(desc(followers))
                                        
head(sorted_data_artists_by_popularity, 10)
head(sorted_data_artists_by_followers, 10)

filtered_data_by_genres <- spotify_artists %>% 
                            filter('k-pop' %in% genres) %>% 
                             arrange(desc(popularity))
head(filtered_data_by_genres, 15) 
```
