```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = 100
freq = 3
x = np.arange(samples)
y1 = np.sin(2 * freq * np.pi * (x / samples))
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
ax.stem(x, y1, 'r', )
ax.plot(x, y1)
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.set_title('Sin wave')
```

```python
def fft_plot(audio, sampling_rate):
    n = len(audio)
    T = 1 / sampling_rate
    yf = scipy.fft.fft(audio)
    xf = np.linspace(0.0, 1.0 / (2.0 * T), int(n / 2))
    
    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111)
    ax.plot(xf, 2 / n * np.abs(yf[:n // 2]))
    ax.grid(True)
    ax.set_xlabel('Frequency')
    ax.set_ylabel('Magnitude')
```

```python
fft_plot(y1, samples)
```

```python
freq = 11
samples = 100
y2 = 2 * np.sin(2 * freq * np.pi * (x / samples))
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
ax.stem(x, y2, 'r', )
ax.plot(x, y2)
ax.set_xlabel('Time')
ax.set_ylabel('Amplitute')
```

```python
fft_plot(y2, samples)
```

```python
y3 = y1 + y2
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
ax.plot(x, y3)
ax.grid(True)
ax.set_xlabel('Time')
ax.set_ylabel('Amplitute')
fft_plot(y3, samples)
```
