```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('avocado.csv')
dataframe.head()
```

```python
main_df = dataframe.copy()
main_df = main_df.drop(columns=['Unnamed: 0'])
```

```python
main_df.describe()
```

```python
main_df.head()
```

```python
dataframe_albany = main_df[main_df['region'] == 'Albany']
dataframe_albany.head()
```

```python
total = pd.value_counts(main_df['region'])

fig = plt.figure(figsize=(10, 15))
ax = fig.add_subplot(111)
sns.barplot(x=list(total), y=total.index)
```

```python
dataframe_albany['month'] = pd.DatetimeIndex(dataframe_albany['Date']).month
```

```python
total_by_months = pd.value_counts(dataframe_albany['month'])

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
sns.barplot(x=total_by_months.index, y=list(total_by_months))
ax.grid(True)
ax.set_xlabel('Months')
ax.set_ylabel('Total count of data')
```
