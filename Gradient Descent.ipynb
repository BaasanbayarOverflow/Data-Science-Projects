```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def gradient_descent(gradient, start, learning_rate, iteration=50):
    vector = start
    vector_points = []
    for _ in range(iteration):
        diff = -learning_rate * gradient(vector)
        vector_points.append(vector)
        vector += diff
    return vector, vector_points
```

```python
vector, points = gradient_descent(gradient=lambda x: x * 2, start=10.0, learning_rate=0.2)
vector
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 2 
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y)
ax.plot(points, np.array(points) ** 2 , '-o', color='red')
```
