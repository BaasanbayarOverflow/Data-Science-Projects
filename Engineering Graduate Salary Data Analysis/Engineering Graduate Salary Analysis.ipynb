```python
df = df.replace(-1, np.nan)
cols_with_nan = [col for col in df.columns if df.isna().sum()[col]>0]
for col in cols_with_nan:
    df[col] = df[col].fillna(df[col].mean())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df.Gender = label_encoder.fit_transform(df.Gender)
df.Specialization = label_encoder.fit_transform(df.Specialization)
df.Degree = label_encoder.fit_transform(df.Degree)
```

```python
df.head(3)
```

```python
df.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_dataframe = df[['Gender', 'Degree', 'Specialization', 'collegeGPA', 'Logical', 'agreeableness', 'openess_to_experience']]
Y_dataframe = df['Salary']

x_train, x_test, y_train, y_test = train_test_split(X_dataframe, Y_dataframe, test_size=0.25, shuffle=True)

linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
```

```python
print(f"Cofficient of determination: {linear_regression.score(x_train, y_train)}")
print(f"{linear_regression.coef_}")
print(f"{linear_regression.intercept_}")
```

```python
y_pred = linear_regression.predict(x_test)
y_pred = np.round(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.plot(y_test, y_test, color='blue')
ax.plot(y_pred, y_pred, color='red')
ax.legend(['Real data', 'Predicted data'])
```

```python
x_data = df.collegeGPA
y_data = df.Salary

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data)
x_train = np.array(x_train).reshape(-1, 1)
x_test = np.array(x_test).reshape(-1, 1)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(f"Cofficient of determination: {model.score(x_train, y_train)}")
print(f"{model.coef_}")
print(f"{model.intercept_}")
```

```python
y_pred = model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
ax.scatter(y_test, y_test, color='red', marker='.')
ax.scatter(y_pred, y_pred, color='blue', marker='o')
ax.legend(['Real values', 'Predicted values'])
```
